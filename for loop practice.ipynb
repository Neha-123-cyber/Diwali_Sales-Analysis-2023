```python
for a in range (10,0,-1):
    print("2*",a,"=",2*a)
```

```python
# Print numbers from 1 to 100.
i=1
while i<=100:
    print(i)
    i+=1
```

```python
## Print numbers from 100 to 1.
i =100
while i>=1:## stopping condition
    print(i)
    i-=1
```
